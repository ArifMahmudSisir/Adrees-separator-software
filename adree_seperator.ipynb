```python
import pandas as pd

# Read the CSV file
input_file = '/content/Contact Information (Responses) - Form responses 1.csv'
output_file = 'processed_addresses.csv'

# Load the data
df = pd.read_csv(input_file)

# Function to extract city, zip code, and country (assuming USA)
def extract_address_parts(Address):
    try:
        parts = Address.split(', ')
        if len(parts) != 3:
            return pd.Series([None, None, None])
        city = parts[0]
        zip_country = parts[1].split(' ')
        if len(zip_country) != 2 or zip_country[1] != 'USA':
            return pd.Series([None, None, None])
        zip_code = zip_country[0]
        country = zip_country[1]
        return pd.Series([city, zip_code, country])
    except Exception as e:
        print(f"Error processing address '{address}': {e}")
        return pd.Series([None, None, None])

# Apply the function to the Address column
df[['City', 'ZipCode', 'Country']] = df['Address'].apply(extract_address_parts)

# Drop the original Address column if not needed
df.drop(columns=['Address'], inplace=True)

# Save the DataFrame to a new CSV file
df.to_csv(output_file, index=False)

print(f"Processed data saved to {output_file}")
```

```python
import pandas as pd

# Read the CSV file
input_file = '/content/Contact Information (Responses) - Form responses 1.csv'
output_file = 'processed_data.csv'

# Function to extract city and zip code from address
def extract_city_zip(row):
    try:
        data = row['Data']
        # Check if the expected splitting pattern is present
        if '," ' not in data:
            raise ValueError(f"Unexpected format: '{data}'")

        parts = data.split('," ')[1].split(', ')
        if len(parts) < 3:
            raise ValueError(f"Insufficient parts after split: '{data}'")

        city = parts[1]
        state_zip = parts[2].split(' ')
        if len(state_zip) < 2:
            raise ValueError(f"Insufficient parts for state and zip: '{data}'")

        zip_code = state_zip[1]

        return pd.Series([city, zip_code])

    except Exception as e:
        print(f"Error processing data '{row}': {e}")
        return pd.Series([None, None])

# Read the input CSV into a DataFrame
df = pd.read_csv(input_file)

# Apply the function to extract city and zip code
df[['City', 'ZipCode']] = df.apply(extract_city_zip, axis=1)

# Drop the original 'Data' column if not needed
df.drop(columns=['Data'], inplace=True)

# Save the DataFrame to a new CSV file
df.to_csv(output_file, index=False)

print(f"Processed data saved to {output_file}")
```

```python
import pandas as pd
import usaddress
## worked this
# Input and output files
input_file = '/content/Contact Information (Responses) - Form responses 1 (1).csv'
output_file = 'standardized_addresses.csv'

# Function to parse and standardize address
def standardize_address(address):
    try:
        parsed_address, address_type = usaddress.tag(address)

        if address_type != 'Street Address':
            raise ValueError(f"Not a valid street address: {address}")

        # Extracting individual components
        components = {
            'AddressNumber': parsed_address.get('AddressNumber', ''),
            'StreetName': parsed_address.get('StreetName', ''),
            'StreetNamePostType': parsed_address.get('StreetNamePostType', ''),
            'OccupancyType': parsed_address.get('OccupancyType', ''),
            'OccupancyIdentifier': parsed_address.get('OccupancyIdentifier', ''),
            'PlaceName': parsed_address.get('PlaceName', ''),
            'StateName': parsed_address.get('StateName', ''),
            'ZipCode': parsed_address.get('ZipCode', ''),
        }

        return pd.Series(components)

    except usaddress.RepeatedLabelError as e:
        print(f"Error standardizing address '{address}': {e}")
        return pd.Series({
            'AddressNumber': '',
            'StreetName': '',
            'StreetNamePostType': '',
            'OccupancyType': '',
            'OccupancyIdentifier': '',
            'PlaceName': '',
            'StateName': '',
            'ZipCode': ''
        })
    except Exception as e:
        print(f"Error standardizing address '{address}': {e}")
        return pd.Series({
            'AddressNumber': '',
            'StreetName': '',
            'StreetNamePostType': '',
            'OccupancyType': '',
            'OccupancyIdentifier': '',
            'PlaceName': '',
            'StateName': '',
            'ZipCode': ''
        })

# Read the input CSV into a DataFrame
try:
    df = pd.read_csv(input_file)
except Exception as e:
    print(f"Error reading CSV '{input_file}': {e}")
    exit(1)

# Apply the standardization function to the 'Address' column
df = df.join(df['Address'].apply(standardize_address))

# Save the DataFrame with standardized address components to a new CSV file
try:
    df.to_csv(output_file, index=False)
    print(f"Standardized addresses saved to {output_file}")
except Exception as e:
    print(f"Error saving CSV '{output_file}': {e}")
    exit(1)
```
